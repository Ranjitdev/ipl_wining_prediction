### Extracting Runs per match

```python
runs_per_match = pd.DataFrame(raw_data1.groupby(['ID', 'BattingTeam', 'innings'], sort=False)['total_run'].sum()).reset_index()
runs_per_match.head(2)
```

```python
innings1 = pd.DataFrame(runs_per_match[runs_per_match['innings']==1])
```

```python
innings2 = pd.DataFrame(runs_per_match[runs_per_match['innings']==2])
```

```python
new_df = pd.merge(left=innings1, right=innings2, on='ID')
new_df.drop(columns=['innings_x', 'innings_y'], inplace=True, axis=1)
new_df.columns = (['ID', 'Team 1', 'first innings', 'Team 2', 'second innings'])
new_df.head(2)
```

```python
for i in range(len(new_df)):
    new_df['winners run']=0
```

```python
for i in range(len(new_df)):
    new_df.iloc[i, 5] = sorted([new_df.iloc[i, 2], new_df.iloc[i, 4]])[1]
```

```python
new_df.head(2)
```

```python
final_data = pd.merge(left=raw_data, right=new_df, on='ID')
final_data.head(5)
```

```python
final_data['Winner'] = final_data['WinningTeam']
final_data.drop(columns=['Team1', 'Team2', 'WinningTeam'], inplace=True, axis=1)
```

```python
final_data.columns=('id', 'Season', 'Match', 'Venue', 'Won By', 'Player of the match', 'Team 1 players', 'Team 2 players', 'Team 1', 'Team 1 Score', 'Team 2', 'Team 2 score', 'Winner score', 'Winner')
```

```python
data = pd.read_csv('all_matches.csv')
data.head(2)
```

### Many teams not playing now ipl so fixing with new data

```python
data['Team 1'].unique()
```

Kings XI Punjab named Punjab Kings  
Delhi Daredevils named Delhi Capitals  
Deccan Chargers named Sunrisers Hyderabad  
Rising Pune Supergiant, Rising Pune Supergiants, Pune Warriors, Gujarat Lions, Kochi Tuskers Kerala dont play matches anymore

```python
def change_names(obj: str):
    not_playing = [
    'Rising Pune Supergiant', 
    'Rising Pune Supergiants', 
    'Pune Warriors', 
    'Gujarat Lions', 
    'Kochi Tuskers Kerala'
    ]
    if obj in not_playing:
        return np.nan
    elif obj == 'Kings XI Punjab':
        return 'Punjab Kings'
    elif obj == 'Delhi Daredevils':
        return 'Delhi Capitals'
    elif obj == 'Deccan Chargers':
        return 'Sunrisers Hyderabad'
    else:
        return obj
```

```python
data['Team 1'] = data['Team 1'].apply(change_names)
data['Team 2'] = data['Team 2'].apply(change_names)
data['Winner'] = data['Winner'].apply(change_names)
data.isna().sum()
```

```python
data.dropna(inplace=True, axis=0)
data.shape
```

```python
data.tail(2)
```

```python
raw_data1['current_run'] = raw_data1.groupby('ID').cumsum()['total_run']
```

```python
over_wise_run = pd.DataFrame(raw_data1.groupby(['ID', 'BattingTeam', 'innings', 'overs'], sort=False)['total_run'].sum()).reset_index()
over_wise_run.to_csv
```

```python
overs_played = pd.DataFrame(over_wise_run.groupby(['ID', 'BattingTeam', 'innings'], sort=False)['overs'].max()).reset_index()
```

```python
overs_first = pd.DataFrame(overs_played[overs_played['innings']==1][['ID', 'overs']])
overs_first.columns=['id', 'Team 1 over']
```

```python
overs_second = pd.DataFrame(overs_played[overs_played['innings']==2][['ID', 'overs']])
overs_second.columns=['id', 'Team 2 over']
```

```python
final_df = pd.merge(left=data, right=overs_first, on='id')
final_df = pd.merge(left=final_df, right=overs_second, on='id')
```

```python
final_df['Team 1 rr'] = final_df['Team 1 Score']/final_df['Team 1 over']
final_df['Team 2 rr'] = final_df['Team 2 score']/final_df['Team 2 over']
final_df.head()
```

```python
final_df=final_df[['id', 'Season', 'Match', 'Venue', 'Won By', 'Player of the match', 'Team 1', 'Team 1 Score', 'Team 1 over', 'Team 1 rr', 'Team 1 players', 'Team 2', 'Team 2 score', 'Team 2 over', 'Team 2 rr', 'Team 2 players', 'Winner', 'Winner score']]
```
